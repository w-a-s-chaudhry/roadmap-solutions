# Data Visualization Tutorial

**Dataset:** Tips — loaded via `sns.load_dataset("tips")`  
**Goal:** Build five chart types (histogram, bar, box, scatter,
heatmap) using both Matplotlib and Seaborn, apply styling,
and export at least one chart as a PNG.

## C1 — Setup & Initial Exploration

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Setup imports, load tips dataset,
#              and run initial exploration.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = sns.load_dataset("tips")

print(df.shape)
df.info()
display(df.describe())
display(df.head())
```

## C2 — Histogram: `total_bill` Distribution

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Histogram of total_bill — Matplotlib vs Seaborn
#              side by side.

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].hist(df["total_bill"], bins=20)
ax[0].set_title("Matplotlib — total_bill Distribution")
ax[0].set_xlabel("Total Bill ($)")
ax[0].set_ylabel("Count")

sns.histplot(df["total_bill"], bins=20, kde=False, ax=ax[1])
ax[1].set_title("Seaborn — total_bill Distribution")
ax[1].set_xlabel("Total Bill ($)")
ax[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

## C3 — Bar Chart: Average Tip by Day

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Bar chart of average tip per day —
#              Matplotlib vs Seaborn.

avg_tip = df.groupby("day")["tip"].mean()

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# --- Matplotlib ---
ax[0].bar(avg_tip.index, avg_tip.values)
ax[0].set_title("Matplotlib \u2014 Average Tip by Day")
ax[0].set_xlabel("Day")
ax[0].set_ylabel("Average Tip ($)")
ax[0].set_ylim(0, avg_tip.max() * 1.2)

# --- Seaborn ---
sns.barplot(
    data=df, x="day", y="tip", estimator="mean", ax=ax[1],
)
ax[1].set_title("Seaborn \u2014 Average Tip by Day")
ax[1].set_xlabel("Day")
ax[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

## C4 — Box Plot: Tip Distribution by Day

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Box plot of tip by day —
#              Matplotlib vs Seaborn.

days = ["Thur", "Fri", "Sat", "Sun"]
tip_by_day = [
    df.loc[df["day"] == d, "tip"].values for d in days
]

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# --- Matplotlib ---
ax[0].boxplot(tip_by_day, labels=days)
ax[0].set_title("Matplotlib \u2014 Tip by Day")
ax[0].set_xlabel("Day")
ax[0].set_ylabel("Tip ($)")

# --- Seaborn ---
sns.boxplot(data=df, x="day", y="tip", ax=ax[1])
ax[1].set_title("Seaborn \u2014 Tip by Day")
ax[1].set_xlabel("Day")
ax[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

## C5 — Scatter Plot: Total Bill vs Tip

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Scatter plot of total_bill vs tip —
#              Matplotlib vs Seaborn.

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# --- Matplotlib ---
ax[0].scatter(df["total_bill"], df["tip"], alpha=0.6)
ax[0].set_title("Matplotlib \u2014 Total Bill vs Tip")
ax[0].set_xlabel("Total Bill ($)")
ax[0].set_ylabel("Tip ($)")

# --- Seaborn ---
sns.scatterplot(
    data=df, x="total_bill", y="tip",
    hue="time", alpha=0.7, ax=ax[1],
)
ax[1].set_title("Seaborn \u2014 Total Bill vs Tip")
ax[1].set_xlabel("Total Bill ($)")
ax[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

## C6 — Correlation Heatmap

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Correlation heatmap of numeric columns —
#              Matplotlib vs Seaborn.

corr = df.select_dtypes("number").corr()

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# --- Matplotlib ---
im = ax[0].imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)
fig.colorbar(im, ax=ax[0])
ax[0].set_xticks(range(len(corr.columns)))
ax[0].set_yticks(range(len(corr.columns)))
ax[0].set_xticklabels(
    corr.columns, rotation=45, ha="right",
)
ax[0].set_yticklabels(corr.columns)
ax[0].set_title("Matplotlib \u2014 Correlation Matrix")

# --- Seaborn ---
sns.heatmap(
    corr, annot=True, fmt=".2f",
    cmap="coolwarm", vmin=-1, vmax=1, ax=ax[1],
)
ax[1].set_title("Seaborn \u2014 Correlation Matrix")

plt.tight_layout()
plt.show()
```

## C7 — Export PNG

```python
# Author:      Wajid Ali Saleem Chaudhry
# Description: Re-draw the Seaborn heatmap and save it
#              to outputs/heatmap.png.

corr = df.select_dtypes("number").corr()

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(
    corr, annot=True, fmt=".2f",
    cmap="coolwarm", vmin=-1, vmax=1, ax=ax,
)
ax.set_title("Correlation Matrix \u2014 Tips Dataset")

plt.tight_layout()
fig.savefig(
    "outputs/heatmap.png", dpi=150, bbox_inches="tight",
)
print("Saved to outputs/heatmap.png")
```
